```python
stuperf_raw=pd.read_csv('/content/drive/MyDrive/StudentsPerformance_raw.csv')
stuperf_raw
```

```python
#Changing the categorical values into numerical class values
stuperf_raw['gender'].replace(['female', 'male'],
                        [0, 1], inplace=True)
stuperf_raw['test preparation course'].replace(['none', 'completed'],
                       [0, 1], inplace=True)

#Creating a new column total score
stuperf_raw['Final score'] = stuperf_raw['math score'] + stuperf_raw['writing score'] + stuperf_raw['reading score']
stuperf_raw
```

```python
stuperf_clean=stuperf_raw.dropna()
stuperf_clean
```

```python
stuperf_clean.to_csv('/content/drive/MyDrive/Stuperf_clean_data.csv')
```
