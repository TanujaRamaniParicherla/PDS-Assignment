```python
frailty_analysis=pd.read_csv('/content/drive/MyDrive/frailty_clean_data.csv', index_col=0)
frailty_analysis
```

```python
# Generate the scatter plot
frailty_analysis_1 = px.scatter(frailty_analysis,
                 x='Height(in)',
                 y='Weight(lb)',
                 color='Frailty',
                 size='Age',
                 hover_data=['Grip_strength(kg)'],
                 title="Height vs. Weight (Colored by Frailty, Sized by Age)")

# Show the plot
frailty_analysis_1.show()
pio.write_image(frailty_analysis_1, '/content/drive/MyDrive/distributionforagewithothers.png')

```

```python
# Box plot
#to check difference in grip strength between frail and non-frail individuals.
frailty_analysis_2 = px.box(frailty_analysis,
             x='Frailty',
             y='Grip_strength(kg)',
             title="Grip Strength Distribution by Frailty",
             labels={'Frailty': 'Frailty (0 = Non-Frail, 1 = Frail)', 'Grip_strength(kg)': 'Grip Strength (kg)'})

frailty_analysis_2.show()
pio.write_image(frailty_analysis_2, '/content/drive/MyDrive/gripstrength.png')
```
