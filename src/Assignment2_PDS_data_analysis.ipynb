b) Remove the units from some of the attributes and only keep the numerical values (for example remove kmpl from “Mileage”, CC from “Engine”, bhp from “Power”, and lakh from “New_price”).

```python
car_data=pd.read_csv('cleaned_data.csv')
car_data
```

```python
car_data['Mileage'] = car_data['Mileage'].str.replace('kmpl', '').str.replace('km/kg', '').astype(float)
car_data['Engine'] = car_data['Engine'].str.replace('CC', '').astype(float)
car_data['Power'] = car_data['Power'].str.replace('bhp', '').astype(float)
car_data['New_Price'] = car_data['New_Price'].str.replace('Lakh', '').str.replace('Cr', '').astype(float)
```

```python
car_data
```

C) Change the categorical variables (“Fuel_Type” and “Transmission”) into numerical one hot encoded value.

```python
car_cat_data = pd.get_dummies(car_data, columns=['Fuel_Type', 'Transmission'])
```

```python
car_cat_data
```

```python
car_cat_data.to_csv('car_data_encoded.csv', index=False)
```

d) Create one more feature and add this column to the dataset (you can use mutate function in R for this). For example, you can calculate the current age of the car by subtracting “Year” value from the current year


```python
import datetime
```

```python
car_data_lat=pd.read_csv('car_data_encoded.csv')
```

```python

current_year = datetime.datetime.now().year
```

```python
car_data_lat['Current_Age'] = current_year - car_data_lat['Year']
```

e) Perform select, filter, rename, mutate, arrange and summarize with group by operations (or their equivalent operations in python) on this dataset.

select

```python

select_car_data_lat = car_data_lat[['Name', 'Location', 'Year', 'Price']]
select_car_data_lat.head()
```

filter

```python

filter_car_data_lat = car_data_lat[(car_data_lat['Year'] == 2015) & (car_data_lat['Owner_Type'] == 'First')]

filter_car_data_lat
```

Rename

```python

rename_car_data_lat = car_data_lat.rename(columns={'Kilometers_Driven': 'KM_Driven'})
rename_car_data_lat.head()
```

 mutate

```python

rename_car_data_lat['Price_per_KM'] = rename_car_data_lat['Price'] / rename_car_data_lat['KM_Driven']
rename_car_data_lat[['Name', 'KM_Driven', 'Price', 'Price_per_KM']].head()
```

```python
rename_car_data_lat
```

```python
rename_car_data_lat.to_csv('final_car_data.csv', index=False)
```

Arrange

```python

sorted_car_data_lat = rename_car_data_lat.sort_values(by='Price', ascending=False)
sorted_car_data_lat[['Name', 'Price']].head()
```

Summarize

```python
summary_car_data_lat = rename_car_data_lat.groupby('Location').agg(average_price=('Price', 'mean'))
summary_car_data_lat
```
