```python
import pandas as pd
```

```python
raw_train_data=pd.read_csv("train.csv")
raw_train_data
```

Cleaning Data

```python
# dropping the unnamed column from the dataset
raw_train_data.drop(raw_train_data.columns[0], axis=1, inplace=True)
raw_train_data
```

a)Look for the missing values in all the columns and either impute them (replace with mean, median, or mode) or drop them. Justify your action for this task

```python

missing_values = raw_train_data.isna().sum()
missing_values
```

```python
# Loop through each column with missing values
for column in missing_values.index:
    if missing_values[column] > 0:
        if raw_train_data[column].dtype == 'object':
            # Replace missing values with the mode for categorical columns
            mode_value = raw_train_data[column].mode()[0]
            raw_train_data[column].fillna(mode_value, inplace=True)
        else:
            # Replace missing values with the mean for numerical columns
            mean_value = raw_train_data[column].mean()
            raw_train_data[column].fillna(mean_value, inplace=True)

# Check if all missing values have been handled
print("Remaining missing values after imputation:")
print(raw_train_data.isnull().sum())
```
